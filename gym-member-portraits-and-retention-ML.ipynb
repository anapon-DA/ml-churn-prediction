# Выявление и анализ факторов, влияющих на отток клиентов сети фитнес-центров, прогнозирование оттока и кластеризация клиентов с применением Machine Learning

Сеть фитнес-центров разрабатывает стратегию взаимодействия с клиентами.
Необходимо провести анализ и подготовить план действий по удержанию клиентов:
- составить прогноз вероятности оттока (на уровне следующего месяца) для каждого клиента;

- сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;

- проанализировать основные признаки, наиболее сильно влияющие на отток;

- сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами: выделить целевые группы клиентов, предложить меры по снижению оттока и определить другие особенности взаимодействия с клиентами.

**Описание данных**

Датасет `gym_churn.csv`, путь к файлу: `/datasets/gym_churn.csv`

Данные клиента за предыдущий до проверки факта оттока месяц:
- `gender` — пол;
- `Near_Location` — проживание или работа в районе, где находится фитнес-центр;
- `Partner` — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);
- `Promo_friends` — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента);
- `Phone` — наличие контактного телефона;
- `Age` — возраст;
- `Lifetime` — время с момента первого обращения в фитнес-центр (в месяцах).

Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:
- `Contract_period` — длительность текущего действующего абонемента (месяц, 6 месяцев, год);
- `Month_to_end_contract` — срок до окончания текущего действующего абонемента (в месяцах);
- `Group_visits` — факт посещения групповых занятий;
- `Avg_class_frequency_total` — средняя частота посещений в неделю за все время с начала действия абонемента;
- `Avg_class_frequency_current_month` — средняя частота посещений в неделю за предыдущий месяц;
- `Avg_additional_charges_total` — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.
- `Churn` — факт оттока в текущем месяце.

## Модель прогнозирования оттока клиентов

### Разделение данных на обучающую и тестовую выборки

```python
# разделим данные на признаки и переменную
X = df.drop('churn', axis=1)
y = df['churn']
```

```python
# разделим модель на обучающую и валидационную выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Обучение модели

Проведем стандартизацию, чтобы в дальнейшем можно было уверенно интерпретировать весовые коэффициенты моделей.

```python
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)
X_test_st = scaler.transform(X_test)
```

Первый способ - логистическая регрессия:

```python
# согласно документации sklearn, регуляризация для lr по умолчанию l2
# установим вместо неё l1, чтобы при интерпретации весов заодно выявить скоррелированные признаки
lr_model = LogisticRegression(solver='liblinear', penalty='l1', random_state = 0)
lr_model.fit(X_train_st, y_train) # обучение
y_lr_pred = lr_model.predict(X_test_st)
```

Второй способ - случайный лес:

```python
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0) 
rf_model.fit(X_train_st, y_train) # обучение
y_rf_pred = rf_model.predict(X_test_st)
```

### Оценка метрик моделей

```python
print(
    'Accuracy\nLogisticRegression: {:.2f}\tRandomForest: {:.2f}\n'.format(
        accuracy_score(y_test, y_lr_pred), accuracy_score(y_test, y_rf_pred)
    )
)
print(
    'Precision\nLogisticRegression: {:.2f}\tRandomForest: {:.2f}\n'.format(
        precision_score(y_test, y_lr_pred), precision_score(y_test, y_rf_pred)
    )
)
print(
    'Recall\nLogisticRegression: {:.2f}\tRandomForest: {:.2f}\n'.format(
        recall_score(y_test, y_lr_pred), recall_score(y_test, y_rf_pred)
    )
)
print(
    'F1\nLogisticRegression: {:.2f}\tRandomForest: {:.2f}'.format(
        f1_score(y_test, y_lr_pred), f1_score(y_test, y_rf_pred)
    )
)
```

Модель логистической регрессии демонстрирует лучшие метрики по сравнению с моделью случайного леса.

```python
coef = pd.Series(lr_model.coef_[0], index=df.drop('churn', axis=1).columns)
```

```python
coef.sort_values(ascending=False, key=lambda x: abs(x))
```

У нас модель логистической регрессии, построенной по логистической функции:


Большие значения z(x) ведут нас к значению логистической функции, близкому к 1 (отток).

И наоборот, малые значения z(x) ведут нас к значению логистической функции, близкому к 0 (клиент остался).

Мы можем интерпретировать значения весов так: высокая средняя частота посещений за неделю за всё время приближает функцию к единице (отток), однако это влияние нивелируется за счет длительного лайфтайма и/или частых посещений клуба клиентом за предыдущий месяц, которые приближают значение функции к нулю (клиент остался).

Таким образом, интерпретируем коэффициенты так: если клиент в предыдущий месяц посещал клуб редко или не посещал вообще, особенно при условии, что он моложе относительно других клиентов, впервые обратился в клуб недавно и/или раньше приходил часто, это сигнализирует об уходе клиента. 

### Вывод

**Основные** признаки, говорящие об уходе клиента, согласно построенной модели логистической регрессии (ML):
- в предыдущий месяц клиент не посещал клуб или делал это редко;
- клиент впервые обратился в клуб недавно;
- средняя частота посещений в неделю за всё время с начала действия абонемента в целом выше, чем за последний месяц (вывод на основании весов коэффициентов).

## Кластеризация клиентов

### Стандартизация данных

Стандартизуем весь датафрейм (ранее стандартизация была проведена только для признаков X).

```python
# отделим столбец с оттоком от остальных данных
df_without_churn = df.drop('churn', axis=1)
```

```python
scaler = StandardScaler()
df_without_churn_st = scaler.fit_transform(df_without_churn)
```

### Дендрограмма на основе матрицы расстояний

```python
linked = linkage(df_without_churn_st, method = 'ward')
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Иерархическая кластеризация', size=14)
plt.show()
```

Дендрограмма предлагает, согласно цветовой палитре, разделение на 4 кластера. Однако по ТЗ необходимо будет выделить 5 кластеров.

### K-Means

Согласно ТЗ, нам необходимо выделить 5 кластеров клиентов.

```python
km = KMeans(n_clusters=5, random_state=0) # по ТЗ задаём число кластеров, равное 5, фиксируем значение random_state
labels = km.fit_predict(df_without_churn_st) 
```

Добавим в таблицу столбец с номерами кластеров:

```python
df['cluster'] = labels
```

```python
for cluster_number in range(5):
    print(f'Кластер {cluster_number}: {len(df.query("cluster == @cluster_number"))} клиентов')
```

Самый многочисленный кластер - кластер 3, самый малочисленный - кластер 1.

### Средние значения признаков для различных кластеров

```python
df.groupby('cluster').agg('mean').T
```

Удобнее будет построить тепловую карту. Для наглядности тепловой карты отсортируем признаки, распределив их на бинарные и количественные, чтобы построить два разных хитмэпа.

```python
features = df.groupby('cluster').agg('mean')
binaries = []
quantitative = []
more_than_12 = []
for col_name in features.columns:
    if df[col_name].max() <= 1:
        binaries.append(col_name)
    elif features[col_name].max() > 12: # 12 - максимум для абонемента, средние значения свыше 12 есть только у возраста и трат
        more_than_12.append(col_name)
    else:
        quantitative.append(col_name)
```

```python
print(binaries)
print(quantitative)
print(more_than_12)
```

```python
plt.figure(figsize=(5, 5))
sns.heatmap(features[binaries].T, annot=True, linewidths=0.5)
plt.title('Бинарные признаки')
plt.show()
```

- пол: явных различий среди кластеров нет
- проживание или работа в районе, где находит клуб, однозначно не характерна для кластера 2 и однозначно характерна для кластеров 3, 4 и 0
- сотрудниками компании-партнера часто являются клиенты из кластера 0
- характерно быть первоначально записанными в рамках акции "приведи друга" для кластера 0, не характерно для кластера 2
- для клиентов кластера 1 нет контактных телефонов
